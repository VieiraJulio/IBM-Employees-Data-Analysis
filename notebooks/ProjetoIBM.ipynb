```python
colunas_nao_ordenadas = [
    'BusinessTravel',
    'Department',
    'EducationField',
    'Gender',
    'JobRole',
    'MaritalStatus',
    'OverTime'  
]
colunas_categoricas_ordenadas = [
    'Education',
    'JobSatisfaction',
    'JobInvolvement',
    'JobLevel',
    'PerformanceRating',
    'RelationshipSatisfaction',
    'StockOptionLevel',
    'WorkLifeBalance'
]
colunas_alvo = ['Attrition']

colunas_numericas = [
    coluna for coluna in df.columns
    if coluna not in (colunas_categoricas_ordenadas + colunas_nao_ordenadas + colunas_alvo)
]
```

```python
df[colunas_numericas].info()
```

```python
df.info()
```

```python
for coluna in colunas_nao_ordenadas:
    df[coluna] = df[coluna].astype("category")
```
